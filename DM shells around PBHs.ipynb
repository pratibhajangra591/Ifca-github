```python
# The evolution of the radius of DM shells around a PBH (of mass, M_PBH=100 solar mass) 
# as a function of time, for different values of the initial radius of the shell demonstrating
# the different regimes: immediate in-fall, turn around and non-stop expansion)

#Reference research paper "WIMPs and stellar-mass primordial black holes are incompatible".

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.integrate import odeint


G= 6.67e-11                # in units of m^3⋅kg^−1⋅s^−2
M_PBH= 1.989e32            # in units of kg
t_i=2.358e5                # in units of seconds
t_eq=1.59246e12            # in units of seconds
 
def j(n,p):
    return [n[1], (((-G*M_PBH)/((n[0]**2))) - (n[0]/(4*p**2)))]

z0= [1e10,2.12044e4]   #with initial radius, r_i=1e10m

ps= np.linspace(2.358e5,2.4e5,2000000) 
ns= odeint(j,z0,ps)
zs= ns[:,0]
z_max= (np.max(zs))
p_max = ps[np.argmax(zs)]
print("r_ta=",'%.4e'%z_max,"m",",","t_ta=",'%.4e'% p_max,"s")






plt.figure(figsize=(12, 6))
plt.rc('lines', linewidth=1.3)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)

plt.plot(ps,zs,'r')
plt.xlabel('$t$ $\hspace{0.5}$[s]',fontsize=12.5)
plt.ylabel('$r$$\hspace{0.5}$ [m]',fontsize=12.5)
plt.title('Dynamics of a DM shell around a PBH of mass, $M_{PBH}=100\hspace{0.3} M_{☉}$ with initial radius, $r_{i}=10^{10}\hspace{0.3}$m')
plt.show()
```

```python
# Dynamics of the various shells of DM halo around a PBH of mass, M_PBH=100 
# solar mass at different initial radii

A=1.495e11                 # Astronomical unit
t_ly=3.154e7               # in units of seconds
G= 6.67e-11                # in units of m^3⋅kg^−1⋅s^−2
M_PBH= 1.989e32            # in units of kg
t_i=2.358e5                # in units of seconds
t_eq=1.59246e12            # in units of seconds
T_eq=(t_eq/t_ly)

def j(n,p):
    return [n[1], (((-G*M_PBH)/((n[0]**2))) - (n[0]/(4*p**2)))]
z0= [((1e10) ),(2.12044e4  )]                      #with initial distance r_i=1e10m
ps= np.linspace((2.358e5),(2.4e5),100000) 
ns= odeint(j,z0,ps)
zs= ns[:,0]
ps_a=ps/t_ly
zs_a=zs/A

def f(u,w):
    return [u[1], (((-G*M_PBH)/((u[0]**2))) - (u[0]/(4*w**2)))]
r0= [1e11,2.12044e5]        #with initial distance r_i=1e11m
ws= np.linspace(2.358e5,7.3e5,100000)
us= odeint(f,r0,ws)
rs= us[:,0]
ws_a=ws/t_ly
rs_a=rs/A

def F(U,W):
    return [U[1], (((-G*M_PBH)/((U[0]**2))) - (U[0]/(4*W**2)))]
R0= [5.8e11,(5.8e11/(2*t_i))]        #with initial distance r_i=5.8e11m
Ws= np.linspace(2.358e5,3.99e8,100000)
Us= odeint(F,R0,Ws)
Rs= Us[:,0]
Rs_a=Rs/A
Ws_a=Ws/t_ly


def h(m,k):
    return [m[1], (((-G*M_PBH)/((m[0]**2))) - (m[0]/(4*k**2)))]                                          
y0= [1e12,2.12044e6]             #with initial distance r_i=1e12m
ks= np.linspace(2.358e5,8.5e9,100000)
ms= odeint(h,y0,ks)
ys= ms[:,0]
ys_a=ys/A
ks_a=ks/t_ly


def g(v,t):
    return [v[1], (((-G*M_PBH)/((v[0]**2))) - (v[0]/(4*t**2)))]
x0= [2.0e+12 ,(2.0e+12/(2*t_i))]            #with initial distance r_i=2.0e+12m
ts= np.linspace(2.358e5,5e11,100000) 
vs= odeint(g,x0,ts)
xs= vs[:,0] 
xs_a=xs/A
ts_a=ts/t_ly


def H(M,K):
    return [M[1], (((-G*M_PBH)/((M[0]**2))) - (M[0]/(4*K**2)))]                                          
Y0= [1e13,2.12044e7]             #with initial distance r_i=1e13m
Ks= np.linspace(2.358e5,1.59246e12,100000)
Ms= odeint(H,Y0,Ks)
Ys= Ms[:,0]
Ys_a=Ys/A
Ks_a=Ks/t_ly



def j(n,p):
    return [n[1], (((-G*M_PBH)/((n[0]**2))) - (n[0]/(4*p**2)))]
z0= [1e14,2.12044e8]   #with initial distance r_i=1e14m
Ps= np.linspace(2.358e5,1.59246e12,100000) 
ns= odeint(j,z0,Ps)
Zs= ns[:,0]
Zs_a=Zs/A
Ps_a=Ps/t_ly



def r(t):  # eq.(5) in "WIMPs and stellar-mass primordial black holes are incompatible".
    return ((2*G*M_PBH*(t**(2)))**(1/3))
t= np.linspace(2.358e5,1.59246e12,100000)





plt.figure(figsize=(8, 6))
plt.style.use('classic')
plt.rc('lines', linewidth=1.3)

plt.loglog(Ps_a,Zs_a,'#4B008B')
plt.loglog(Ks_a,Ys_a,'#7E1E9C')
plt.loglog(ts_a,xs_a,'#9A0EEA')
plt.loglog(ks_a,ys_a,'#C20078')
plt.loglog(Ws_a,Rs_a,'#DA70D6')
plt.loglog(ws_a,rs_a,'#FF81C0')
plt.loglog(ps_a,zs_a,'#DDA0DD')

plt.loglog(t/t_ly,r(t)/A,'k:')
plt.axvline(x=T_eq, color='c', linestyle='--')

plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.xlabel('$t$ (years)',fontsize=13)
plt.ylabel('$r$ (AU)',fontsize=13)
plt.legend(['$r_{i}=670\hspace{0.4}AU$','$r_{i}=67$','$r_{i}=10$','$r_{i}=6.7$','$r_{i}=3.8$','$r_{i}=0.7$','$r_{i}=0.06$','$r_\mathrm{turnaround}$','$t_\mathrm{eq}=50490\hspace{0.4} years$'],prop = {'size':7.6},
              loc=(0.03,0.63),shadow = True,
           facecolor = 'white',fontsize=44)
plt.savefig("DM shells around PBH.pdf")
plt.show()
```
