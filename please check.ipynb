```python
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.integrate import odeint


G=6.67e-11 # with units m^3 kg^-1 s^-1
M_PBH=1.989e32 #in kg
 
r_i=1e11
t_i=2.358e5
while r_i < 1e14:
    t_ta=((r_i**6)/(128*((G*M_PBH)**2)*(t_i**3)))
    print("t_ta =",t_ta,"s")
    def f(u,t):
         return [u[1], (((-G*M_PBH)/((u[0]**2))) - (u[0]/(4*t**2)))]
    r0= [r_i,(r_i/(2*t_i))]
    ts= np.linspace(t_i,(2.19*t_ta),100000)
    us= odeint(f,r0,ts)
    rs= us[:,0]
    
    plt.plot(ts,rs,'g')
    plt.xlabel('t in seconds')
    plt.ylabel('r in meters')
    plt.title('')
    plt.legend(["","",""])
    r_i*=10
plt.show()
```
