```python
#the evolution of an equal-mass PBH binary with a time-dependent DM halo. 

import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.integrate import odeint

def h(x):
    return math.sqrt((x**-3)+(x**-4))
def g(x):
        return  -((3*x+4)/(2*h(x)*(x**5))) #g(x)=h'(x).
rho_eq=1.3911e-16
t_eq=1.997e12
pi=3.14
G=6.67e-11
def n(s):
    return (1+((((64*pi*G*rho_eq*(t_eq**2))/3)**(3/4))*(((((2*(s-2)*((s+1)**(1/2)))/3)+(4/3))**(1/2)))))

def  k(s):
    return (1/n(s))

def λ(s):
    return λ_0*k(s)

def f(u,s):
    return [u[1], -((s*g(s)+h(s))*((s*u[1]-u[0]))/(h(s)*s**2))-((u[0])/((((h(s)*s)**2))*2*λ(s)*((u[0]**2
                                                                                    +b**2)**(3/2))))]
λ_0=1
b=1e-6
separation0= [1e-12,1]
ss= np.linspace(1e-12,5,10000000) 
us= odeint(f,separation0,ss)
separations= us[:,0]
As=ss/λ_0
Bs=separations/λ_0

plt.figure(figsize=(16, 8))

plt.plot(As,Bs,'g')
plt.xlabel('s/λ_0')
plt.ylabel('X/λ_0')
plt.title('Graph for the evolution of an equal-mass PBH binary with a time-dependent DM halo using λ_0=1') 
plt.show()
```
