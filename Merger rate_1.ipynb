```python
import numpy as np
from scipy.integrate import odeint
import math
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit


G=6.67e-11 
rho_m=3.77e19
alpha=1.6
M=5
M_1PBH=5
M_2PBH=5
M_binary=(M_1PBH+M_2PBH)
c=3e8
rho_eq=2.15368e-16
sigma_eq=(0.005)


λ_0=1


A=((G**(9/37))*((M_1PBH*M_2PBH)**(3/37))*(M_binary**(36/37))*((rho_eq**(4/37))/(c**(15/37))))

def C(λ_0):
    return (1+0.0136*λ_0)/((1+0.1709*λ_0)**(1/2))

def F(f,λ_0):
    return ((f**2)*(((f**2)+1.1998*(sigma_eq**2))**(-21/74)))/((1+0.1709*λ_0)**(12/37))

def P(m):
    return (((alpha-1)/M)*((m/M)**(alpha)))   # Powerlaw PBH mass distribution


def minimum(a, b):
      
    if a <= b:
        return a
    else:
        return b     


def merger_rate(f,λ_0):
    return (1e6*A*(C(λ_0))*F(f,λ_0)*((P(M_1PBH)/M_1PBH)+(P(M_2PBH)/M_2PBH))*minimum((P(M_1PBH)/M_1PBH),(P(M_2PBH)/M_2PBH)))
         #eq.(221) in the pdf "Tidal torque".

f=np.linspace(1e-5,1,1000)
f_pbh=(f/0.85)

plt.rc('lines', linewidth=2.4)    
plt.figure(figsize=(8,5))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.yscale('log')

plt.loglog(f_pbh,merger_rate(f,λ_0),'g')
    
    
plt.xlabel('$f_{pbh}$',fontsize=12)
plt.ylabel('$merger\hspace{0.5} rate\hspace{0.5}Gpc^{-3}\hspace{0.5}yr^{-1}$',fontsize=12)
plt.show()
```

```python
rho_eq=2.15368e-16
π=np.pi
sigma=0.6
m_c=15
M_1PBH=5
M_2PBH=5

def P(m):              # Lognormal PBH mass distribution
    return (1/((2*π*sigma*m)**(1/2)))*(math.exp(-((math.log(m/m_c))**2)/(2*(sigma**2))))
    


def merger(f,λ_0):
    return (1e6*A*(C(λ_0))*F(f,λ_0)*((P(M_1PBH)/M_1PBH)+(P(M_2PBH)/M_2PBH))*minimum((P(M_1PBH)/M_1PBH),(P(M_2PBH)/M_2PBH)))
           #eq.(221) in the pdf "Tidal torque".


plt.rc('lines', linewidth=2.4)    
plt.figure(figsize=(8,5))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.yscale('log')



plt.loglog(f_pbh,merger(f,λ_0),'r--')

    
plt.xlabel('$f_{pbh}$',fontsize=12)
plt.ylabel('$merger\hspace{0.5} rate\hspace{0.5}Gpc^{-3}\hspace{0.5}yr^{-1}$',fontsize=12)
plt.show() 
```

```python
plt.rc('lines', linewidth=2.4)    
plt.figure(figsize=(8,5))
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.yscale('log')

plt.loglog(f_pbh,merger_rate(f,λ_0),'g')
plt.loglog(f_pbh,merger(f,λ_0),'r--')

plt.legend(["Powerlaw","Lognormal"])
plt.xlabel('$f_{pbh}$',fontsize=12)
plt.ylabel('$merger\hspace{0.5} rate\hspace{0.5}Gpc^{-3}\hspace{0.5}yr^{-1}$',fontsize=12)
plt.title("Monochromatic PBHs-mass functions")
plt.show()
```
