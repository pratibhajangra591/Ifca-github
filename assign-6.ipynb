```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.integrate import odeint
def f(u,x):
    return [u[1], ((-1/(2*u[0]**2)) - (u[0]/(4*x**2)))]
y0= [3.3920*10e6,7.0681*10e-3]
xs= np.linspace(2.3995*10e8,20.321*10e14,600000)
us= odeint(f,y0,xs)
ys= us[:,0]
plt.plot(xs,ys,'r')
plt.xlabel('t/2GM')
plt.ylabel('r/2GM')
plt.title('using r = 10e12m & M = 100 solar mass')
plt.show()
```
