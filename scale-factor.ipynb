```python
#Task no. 4 For the case of PBH binaries without DM, check the relation between the semi-major axis a 
#and the comoving separation x, which is given in Eq. (11) of https://arxiv.org/abs/1709.06576. 



import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.integrate import odeint



def f(u,s):
    return [u[1], ((s*u[1]-u[0])/(s**2))-((u[0]*(s**2))/(λ*((u[0]**2+b**2)**(3/2))))]
separation0= [1e-12,1]
λ=0.0009
b=1e-6
ss= np.linspace(1e-12,3e-3,1000)  
us= odeint(f,separation0,ss)
separations= us[:,0]
As=ss/λ
Bs=separations/λ
Bs_max= (np.max(Bs))
a= Bs_max/2
print("λ=0.0009","(X/λ)_max=",Bs_max,"a=",a)


def j(n,p):
    return [n[1], ((p*n[1]-n[0])/(p**2))-((n[0]*(p**2))/(λ*((n[0]**2+c**2)**(3/2))))]
z0= [1e-12,1]
λ=0.001
c=1e-6
ps= np.linspace(1e-12,3e-3,3000)
ns= odeint(j,z0,ps)
zs= ns[:,0]
Ps=ps/λ
Zs=zs/λ
Zs_max= (np.max(Zs))
a= Zs_max/2
print("λ=0.001","(X/λ)_max=",Zs_max,"a=",a)

def g(v,t):
    return [v[1], ((t*v[1]-v[0])/(t**2))-((v[0]*(t**2))/(λ*((v[0]**2+d**2)**(3/2))))]
λ=0.003
d= 1e-6
x0= [1e-12,1]
ts= np.linspace(1e-12,1e-2,2500) 
vs= odeint(g,x0,ts)
xs= vs[:,0]
Ts=ts/λ
Xs=xs/λ
Xs_max= (np.max(Xs))
a= Xs_max/2
print("λ=0.003","(X/λ)_max=",Xs_max,"a=",a)


def h(m,k):
    return [m[1],((k*m[1]-m[0])/(k**2))-((m[0]*(k**2))/(λ*((m[0]**2+e**2)**(3/2))))]
λ=0.009
e= 1e-6                                         
y0= [1e-12,1]
ks= np.linspace(1e-12,2e-2,3000)
ms= odeint(h,y0,ks)
ys= ms[:,0]
Ks=ks/λ
Ys=ys/λ
Ys_max= (np.max(Ys))
a= Ys_max/2
print("λ=0.009","(X/λ)_max=",Ys_max,"a=",a)

plt.figure(figsize=(18, 9))

plt.plot(As,Bs,'k')
plt.plot(Ps,Zs,'r')
plt.plot(Ts,Xs,'b')
plt.plot(Ks,Ys,'g')

plt.xlabel('s/λ')
plt.ylabel('X/λ')  
plt.legend(["λ=0.0009","λ=0.001","λ=0.003","λ=0.009"]) 
plt.title('Graph for the solution of eq.(9) with b=1e6 using different λ-parameters ')
# E.(9) of paper 'The merger rate of primordial-black-hole binaries'.
plt.show()
```

```python
# Task no. 5 How does this relation change when we include the (time-dependent) DM halo?


rho_eq=1.3911e-16
t_eq=1.997e8
pi=3.14

def n(s):
    return (1+((((8*pi*rho_eq*(t_eq**2))/3)**(3/4))*(2**(7/4))*s))

def  k(s):
    return (1/n(s))

def λ(s):
    return λ_0*k(s)

def f(u,s):
    return [u[1], ((s*u[1]-u[0])/(s**2))-((u[0]*(s**2))/(λ(s)*((u[0]**2+b**2)**(3/2))))]
λ_0=0.0009
b=1e-6
separation0= [1e-12,1]
ss= np.linspace(1e-12,3e-3,2000)
us= odeint(f,separation0,ss)
separations= us[:,0]
As=ss/λ_0
Bs=separations/λ_0
Bs_max= (np.max(Bs))
a= Bs_max/2
print("λ=0.0009","(X/λ)_max=",Bs_max,"a=",a)


def j(n,p):
    return [n[1], ((p*n[1]-n[0])/(p**2))-((n[0]*(p**2))/(λ(p)*((n[0]**2+c**2)**(3/2))))]
z0= [1e-12,1]
λ_0=0.001
c=1e-6
ps= np.linspace(1e-12,3e-3,2500) 
ns= odeint(j,z0,ps)
zs= ns[:,0]
Ps=ps/λ_0
Zs=zs/λ_0
Zs_max= (np.max(Zs))
a= Zs_max/2
print("λ=0.001","(X/λ)_max=",Zs_max,"a=",a)


def g(v,t):
    return [v[1], ((t*v[1]-v[0])/(t**2))-((v[0]*(t**2))/(λ(t)*((v[0]**2+d**2)**(3/2))))]
λ_0=0.003
d= 1e-6
x0= [1e-12,1]
ts= np.linspace(1e-12,6e-3,2500) 
vs= odeint(g,x0,ts)
xs= vs[:,0]
Ts=ts/λ_0
Xs=xs/λ_0
Xs_max= (np.max(Xs))
a= Xs_max/2
print("λ=0.003","(X/λ)_max=",Xs_max,"a=",a)


def h(m,k):
    return [m[1],((k*m[1]-m[0])/(k**2))-((m[0]*(k**2))/(λ(k)*((m[0]**2+e**2)**(3/2)))) ]
λ_0=0.009
e= 1e-6                                          
y0= [1e-12,1]
ks= np.linspace(1e-12,3e-2,40000)
ms= odeint(h,y0,ks)
ys= ms[:,0]
Ks=ks/λ_0
Ys=ys/λ_0
Ys_max= (np.max(Ys))
a= Ys_max/2
print("λ=0.009","(X/λ)_max=",Ys_max,"a=",a)




plt.figure(figsize=(16, 8))

plt.plot(As,Bs,'k')
plt.plot(Ps,Zs,'g')
plt.plot(Ts,Xs,'b')
plt.plot(Ks,Ys,'r')

plt.xlabel('s/λ_0')
plt.ylabel('X/λ_0')  
plt.legend(["λ=0.0009","λ=0.001","λ=0.003","λ=0.009"]) 
plt.title('Graph for the solution of eq.(78) for with b=1e6 using different λ-parameters ') 
plt.show()
```
