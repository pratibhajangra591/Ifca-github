```python
M=1.989e32

def f(u,s):
    return [u[1], -(((s*u[1]-u[0]))/(s**2))-((u[0]*(M+3.9694e29*s)*((s)**2)))/(      # eq.(50) in the assign-14 pdf.
                   M*位*((u[0]**2+b**2)**(3/2)))]
位=0.009
b=1e-4
separation0= [1e-12,1]
ss= np.linspace(1e-12,1.9e-2,1000) 
us= odeint(f,separation0,ss)
separations= us[:,0]
As=ss/位
Bs=separations/位


def j(n,p):
    return [n[1], -(((p*n[1]-n[0]))/(p**2))-((n[0]*(M+3.9694e29*p)*((p)**2))/(
                   M*位*((n[0]**2+c**2)**(3/2))))]
z0= [1e-12,1]
位=0.009
c=1e-5
ps= np.linspace(1e-12,1.9e-2,10000) 
ns= odeint(j,z0,ps)
zs= ns[:,0]
Ps=ps/位
Zs=zs/位

def w(v,t):
    return [v[1], -(((t*v[1]-v[0]))/(t**2))-((v[0]*(M+3.9694e29*t)*((t)**2))/(
                   M*位*((v[0]**2+d**2)**(3/2))))]
位=0.009
d= 1e-6
x0= [1e-12,1]
ts= np.linspace(1e-12,1.9e-2,10000) 
vs= odeint(w,x0,ts)
xs= vs[:,0]
Ts=ts/位
Xs=xs/位

def i(m,k):
    return [m[1],-(((k*m[1]-m[0]))/(k**2))-((m[0]*(M+3.9694e29*k)*((k)**2))/(
                   M*位*((m[0]**2+e**2)**(3/2))))]
位=0.009
e= 1e-7
y0= [1e-12,1]
ks= np.linspace(1e-12,1.9e-2,1000000) 
ms= odeint(i,y0,ks)
ys= ms[:,0]
Ks=ks/位
Ys=ys/位

plt.figure(figsize=(19, 10))


plt.plot(As,Bs,'m')
plt.plot(Ps,Zs,'r')
plt.plot(Ts,Xs,'b')
plt.plot(Ks,Ys,'g')

plt.xlabel('s/位')
plt.ylabel('X/位')  
plt.legend(["b=1e-4","b=1e-5","b=1e-6","b=1e-7"])  
plt.title('Graph for the solution of eq.(9) modified with variable DM halo mass & 位=0.009 using different impact parameters')  
#  modification of Eq.(9) of paper 'The merger rate of primordial-black-hole binaries' is given by eq.(50) in the assign-14 pdf.
plt.show()
```
