```python
# Interpolation of the scale factor of the binary decouple, s_dec for binaries with and without 
# dark matter halos.


import numpy as  np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from scipy.optimize import curve_fit
import math
from module import *


# Interpolation of the scale factor of the binary decouple, s_dec for binaries without DM halos. 


s_dec_list=np.zeros(len(λ_list))
for i, λ_0 in enumerate(λ_list):
    s_dec_list[i]= s_dec_nohalo(λ_0)

    
    
# Fitting of the the curve for the value of s_dec/λ vs λ for the PBH binaries with and
# without dark matter halos.

x = np.array(λ_list)

y = np.array(s_dec_list)  


z = np.polyfit(x, y, 1)
p = np.poly1d(z)
print("p=",p)



plt.figure(figsize=(8,5))
plt.yscale('log')
plt.loglog(x,y,'go')
plt.plot(x,p(x),'r')
plt.xlabel('$λ$',fontsize=13)
plt.ylabel('$s_{dec}/λ$',fontsize=13)
plt.show()


#Calculation of the Percentage error in the numerical and fitted value of s_dec/λ. 


y_list = y.tolist()
q = np.array(p(x))
q_list = q.tolist()

def percentage_error(numerical, fitted): 
    return (np.abs((numerical - fitted) / numerical)) * 100



error_list=np.zeros(len(λ_list))

for i, (y,q) in enumerate(zip(y_list,q_list)):
    error_list[i] = percentage_error(y, q)    #Percentage error 


error_max=np.max(error_list)
print("Maximum Percentage error= "'%0.4f'%error_max,"%")            
```

```python
# Interpolation of the scale factor of the binary decouple, s_dec for binaries with DM halos. 

s_dec_list=np.zeros(len(λ_list))
for i, λ_0 in enumerate(λ_list):
    s_dec_list[i]= s_dec_halo(λ_0)

    
    
# Fitting of the the curve for the value of s_dec/λ vs λ for the PBH binaries with 
# dark matter halos.

x = np.array(λ_list)

y = np.array(s_dec_list)  


z = np.polyfit(x, y, 1)
p = np.poly1d(z)
print("p=",p)



plt.figure(figsize=(8,5))
plt.yscale('log')
plt.loglog(x,y,'go')
plt.plot(x,p(x),'r')
plt.xlabel('$λ$',fontsize=13)
plt.ylabel('$s_{dec}/λ$',fontsize=13)
plt.show()


#Calculation of the Percentage error in the numerical and fitted value of s_dec/λ. 


y_list = y.tolist()
q= np.array(p(x))
q_list=q.tolist()

def percentage_error(numerical, fitted): 
    return (np.abs((numerical - fitted) / numerical)) * 100



error_list=np.zeros(len(λ_list))

for i, (y,q) in enumerate(zip(y_list,q_list)):
    error_list[i] = percentage_error(y, q)    #Percentage error 


error_max=np.max(error_list)
print("Maximum Percentage error= "'%0.4f'%error_max,"%")
```

```python
# So, the general form of s_dec for PBH binaries with and without dark matter halos 
# is given as:-
               #         s_dec = (0.3356 + 0.08206*λ_0)
                   
               #        s_dec_halo = (0.3425 + 0.03062*λ_0)
```

```python
# Interpolation of the initial semi-major axis, a_i of the PBH binaries with and without 
# dark matter halos.
#Semi-major axis, a=C.λ.x where, C=constant and x=comoving separation of the PBH binary.
# a/(x.λ) is dimensionless.
```

```python
# For PBH binaries without DM halos.

a_list = np.zeros(len(λ_list))    
for k, λ_0 in enumerate(λ_list):
    a_list[k] = semi_nohalo(λ_0)
     
plt.rc('lines', linewidth=2.3) 


plt.figure(figsize = (6,4))    
plt.loglog(λ_list,a_list,'k--')
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.xlabel('$λ$',fontsize = 13)
plt.ylabel('$a/(x\cdotλ)$',fontsize = 13)   # a = semi-major axis of the PBH binary.        
plt.title('PBH binary without DM halos')
plt.show()   



# Fitting of the the curve for a/(x.λ) vs λ for the PBH binary without 
# dark matter halos.


x = np.array(λ_list)

y = np.array(a_list)  


z = np.polyfit(x, y, 1)
p = np.poly1d(z)
print("p =",p)


plt.figure(figsize = (8,5))
plt.yscale('log')
plt.loglog(x,y,'go')
plt.plot(x,p(x),'r')
plt.xlabel('$λ$',fontsize = 13)
plt.ylabel('$a/(x\cdotλ)$',fontsize = 13)
plt.show()


#Calculation of the Percentage error in the numerical and fitted value of semi-major axis, a/(x.λ) 
#for PBH binary without dark matter halos.

y_list = y.tolist()
q = np.array(p(x))
q_list = q.tolist()

def percentage_error(numerical, fitted): 
    return (np.abs((numerical - fitted) / numerical)) * 100



error_list = np.zeros(len(λ_list))

for i, (y,q) in enumerate(zip(y_list,q_list)):
    error_list[i] = percentage_error(y, q)    #Percentage error 


error_max = np.max(error_list)
print("Maximum Percentage error = "'%0.4f'%error_max,"%")
```

```python
# For PBH binaries with dark matter halos.

a_halo_list = np.zeros(len(λ_list))  # a = semi-major axis of the PBH binary.
for k, λ_0 in enumerate(λ_list):
    a_halo_list[k] = semi_halo(λ_0)


    
plt.figure(figsize = (6,4))    
plt.loglog(λ_list,a_halo_list,'C1')
plt.xlabel('$λ$',fontsize = 13) 
plt.ylabel('$a/(x\cdotλ)$',fontsize=13)         
plt.title('PBH binary with DM halos')        
plt.show() 



# Fitting of the the curve for the value of semi-major axis, a/(x.λ) vs λ for the PBH binary with 
# dark matter halos.
x_halo = np.array(λ_list)

y_halo = np.array(a_halo_list)  


z = np.polyfit(x_halo, y_halo, 1)
p_halo = np.poly1d(z)
print("p=",p_halo)


plt.figure(figsize = (8,5))
plt.loglog(x_halo,y_halo,'go')
plt.plot(x_halo,p(x_halo),'r')
plt.xlabel('$λ$',fontsize = 13)
plt.ylabel('$a/(x\cdotλ)$',fontsize = 13)
plt.show()

#This gives semi-major axis, (a_halo/x) = (0.09769λ+0.005742λ^2)m.



#Calculation of the Percentage error in the numerical and fitted value of semi-major axis, a/(x.λ) 
#for PBH binary without dark matter halos.

y_halo_list = y_halo.tolist()
q_halo = np.array(p_halo(x_halo))
q_halo_list = q_halo.tolist()

def percentage_error(numerical, fitted): 
    return (np.abs((numerical - fitted) / numerical)) * 100



error_halo_list = np.zeros(len(λ_list))

for i, (y,q) in enumerate(zip(y_halo_list,q_halo_list)):
    error_halo_list[i] = percentage_error(y, q)    #Percentage error 


error_halo_max = np.max(error_halo_list)
print("Maximum Percentage error = "'%0.4f'%error_halo_max,"%")
```

```python
# So, the general form of (a_i/x) for PBH binaries without dark matter halos 
# is given as:-
   #                (a_i/x) = (0.09647 + 0.01652*λ_0)
   #                (a_i/x)_halo = (0.09766 + 0.006796*λ_0)
```

```python
#Variation of semi-major axis, a/(λ.x) with λ for the PBH binaries with and without 
# DM halos.

plt.rc('lines', linewidth=1.4)
plt.figure(figsize=(9,5))

plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.loglog(λ_list,a_list,'k--')
plt.loglog(λ_list, a_halo_list,'C1')
plt.xlabel('$λ$',fontsize=11)
plt.ylabel('$a/(x\cdotλ)$',fontsize=11)
plt.legend(['$PBH\hspace{0.7} binary\hspace{0.7} without\hspace{0.7} DM\hspace{0.7} halos$','$PBH\hspace{0.7} binary\hspace{0.7} with\hspace{0.7} DM\hspace{0.7} halos$'],prop = {'size' :12},
              loc='upper left',shadow = True,
               facecolor = 'white',fontsize=16)        

plt.title('Variation of a/(x.λ) with λ')
plt.savefig('a_(x.λ)vs λ.pdf')
plt.show()
```
