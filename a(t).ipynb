```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import math


π=np.pi
G=6.67e-11             # in units of m^3.kg^-1.s^-2    
t_eq=1.59246e12        # in units of seconds

A=((8*π*G)/3)

def j(s,t):   # eq.(23) in the attached picture
    return [s[1], -((3/2)*A*s[0]*(((s[0]**(-4))+(s[0]**(-3)))**(2)))]


s0=[0,0]
ts=np.linspace(0,1e15,10) #Since the error was coming, so to chcek for simplicity 
                          #I added 10 points(just for a trial)
ss= odeint(j,s0,ts)
zs= ss[:,0]
a_list=zs.tolist()
print("a=",a_list,"m")


plt.figure(figsize=(6,5))
plt.style.use('classic')
plt.rc('lines',linewidth=2.1)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.plot(ts,zs,'r')
plt.ylabel('$a\hspace{3}[m]$')
plt.xlabel('$t\hspace{3} [s]$')
plt.show()
```
