```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from matplotlib.pyplot import figure
import math


def f(x):
    return (((x**2))/(((x**2)+1)**(3/2)))

def g(x):
    return math.exp(-x)
f2 = np.vectorize(g)

def h(x):
    return f(x)*f2(x)


figure(figsize=(8, 5))
x=np.linspace(0,100,1000)


plt.rc('lines', linewidth=2.4)    

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.loglog(x,h(x),'r')
plt.loglog(x,f(x),'g')
plt.xlabel('x',fontsize=12)
plt.ylabel('h(x)',fontsize=12)
plt.legend(["integrand with exponential factor","integrand without exponential facor"])
plt.show()
```
