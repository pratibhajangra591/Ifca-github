```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from matplotlib.pyplot import figure
import math


def f(x):
    return (((x**2))/(((x**2)+1)**(3/2)))

def g(x):
    return math.exp(-x)
f2 = np.vectorize(g)

def h(x):
    return f(x)*f2(x)


figure(figsize=(8, 5))
x=np.linspace(0,100,1000)


plt.rc('lines', linewidth=2.4)    

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.loglog(x,h(x),'r')
plt.loglog(x,f(x),'g')
plt.xlabel('x',fontsize=12)
plt.ylabel('h(x)',fontsize=12)
plt.legend(["integrand with exponential factor","integrand without exponential facor"])
plt.show()
```

```python
f=0.0043   #say
B=0.001   #say
D=(1/(f*(B**(-21/37))))    # for D to have any value.


def f(y,D):   # eq.(1B) in the notes.
    return ((((1+0.1709*D*(y**(-21/37)))**(-1/14))*y)/(1+(y**(-21/37))*0.0136*D))

def g(y,D):   # eq.(2B) in the notes.
    return ((f(y,D)**2)/(((f(y,D)**2)+1)**(3/2)))*(y**(-58/37))*(math.exp(-((y/B)**(-21/37)))) 

f2 = np.vectorize(g)


def i(y,D):    # eq.(2B) without the exponential factor.
    return ((f(y,D)**2)/(((f(y,D)**2)+1)**(3/2))*(y**(-58/37)))





y=np.linspace(0.0001,1000,100000) 




plt.figure(figsize=(10,5))    
plt.rc('lines', linewidth=2.4)    
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.loglog(y,f2(y,D),'r')
plt.loglog(y,i(y,D),'g')


plt.legend(["integrand with exponential factor","integrand without exponential factor"],prop = {'size' :13},
              loc='upper right',shadow = True,
           facecolor = 'white',fontsize=19)
plt.xlabel('$y$',fontsize=12)
plt.ylabel('$P(y)$',fontsize=12)
plt.show()  
```
