```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.integrate import odeint



def f(u,w):
    return [u[1], (((-2*G*M*u[0])/((u[0]**2+b**2)**(3/2))) - (u[0]/(4*w**2)))]
G= 6.67e-11
M= (1.989e32)
b= 1e9
r0= [1e12,2.12044e6]
ws= np.linspace(2.358e5,2e10,1500)
us= odeint(f,r0,ws)
rs= us[:,0]


def j(n,p):
    return [n[1], (((-2*G*M*n[0])/((n[0]**2+e**2)**(3/2))) - (n[0]/(4*p**2)))]
G= 6.67e-11
M= (1.989e32)
e= 1e10
z0= [1e12,2.12044e6]
ps= np.linspace(2.358e5,2e10,1000)
ns= odeint(j,z0,ps)
zs= ns[:,0]

def g(v,t):
    return [v[1], (((-2*G*M*v[0])/((v[0]**2+c**2)**(3/2))) - (v[0]/(4*t**2)))]
G= 6.67e-11
M= (1.989e32)
c= 1e11
x0= [1e12,2.12044e6]
ts= np.linspace(2.358e5,2e10,2000)
vs= odeint(g,x0,ts)
xs= vs[:,0]


def h(m,k):
    return [m[1], (((-2*G*M*m[0])/((m[0]**2+d**2)**(3/2))) - (m[0]/(4*k**2)))]
G= 6.67e-11
M= (1.989e32)
d= 1e12
y0= [1e12,2.12044e6]
ks= np.linspace(2.358e5,2e10,600)
ms= odeint(h,y0,ks)
ys= ms[:,0]




plt.figure(figsize=(20, 12))

plt.plot(ws,rs,'ko')
plt.plot(ps,zs,'b--')
plt.plot(ts,xs,'g')
plt.plot(ks,ys,'r')
plt.xlabel('t in seconds')
plt.ylabel('r in meters')  
plt.legend(["b=1e9m","b=1e10m","b=1e11m","b=1e12m"])     
plt.show()
```
