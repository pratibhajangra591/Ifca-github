```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.integrate import odeint
def f(u,t):
    return [u[1], (((-2*G*M*u[0])/((u[0]**2+b**2)**(3/2))) - (u[0]/(4*t**2)))]
G= 6.67e-11
M= (1.989e32)
b= 1e10
r0= [1e10,2.12044e4]
ts= np.linspace(2.358e5,1e5,70)
us= odeint(f,r0,ts)
rs= us[:,0]
r_min=(np.min(rs))
r_max= (np.max(rs))
print(r_min,r_max)
print(r_max+r_min)
plt.plot(ts,rs,'m')
plt.xlabel('t in seconds')
plt.ylabel('r in meters')
plt.title(' r = 10e10m & M = 100 solar mass')
plt.show()
```
