```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from matplotlib.pyplot import figure
import math


def f(x):
    return (((x**2))/(((x**2)+1)**(3/2)))

def g(x):
    return math.exp(-x)
f2 = np.vectorize(g)

def h(x):
    return f(x)*f2(x)


figure(figsize=(8, 5))
x=np.linspace(0,100,1000)

plt.plot(x,h(x),'r')
plt.xlabel('x')
plt.ylabel('h(x)')
plt.show()
```
