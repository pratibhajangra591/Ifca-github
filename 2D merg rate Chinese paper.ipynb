```python
import numpy as np
import matplotlib.pyplot as plt
import math
import seaborn as sns

π=np.pi
t_0=13.78e9      # in units of years
t=13.78e9        # in units of years , let t=t_0 (i.e. calcluating the merger rate today)
σ_eq=0.005


m_1=6
m_2=8

y_1=math.log(m_1)
y_2=math.log(m_2)
dy=(y_2-y_1)


y_list=np.zeros(10)

for i in range(len(y_list)):
    y_list[i]= (y_1+(i*dy))
    

M_1_list=np.zeros(len(y_list))
M_2_list=np.zeros(len(y_list))

for i, y in enumerate(y_list):
    M_1_list[i]=(math.ceil(math.exp(y)))
    M_2_list[i]=(math.ceil(math.exp(y)))
    

            
m_1_list=[int(M_1) for M_1 in M_1_list]
m_2_list=[int(M_2) for M_2 in M_2_list]

    
Δ_1_list=np.zeros(len(y_list))
Δ_2_list=np.zeros(len(y_list))

for i, y in enumerate(y_list):
    Δ_1_list[i] = math.exp(y + 0.5*dy)-math.exp(y - 0.5*dy)
    Δ_2_list[i] = math.exp(y + 0.5*dy)-math.exp(y - 0.5*dy) 
        
```

```python
# Calculation of the 2D merger rate for Powerlaw PBH mass distribution. 
#The resolution of the PBHs masses m_1 is Δ_1 and of m_2 is Δ_2 with Δ_1=Δ_2=Δ.

def P(m):     #Powerlaw PBH mass distribution
    α = 1.6
    M=5       #in units of M☉
    return (α-1)/M * (m/M)**(-α)

# merger rate 
def Merger_power(m_1, m_2, t):
    f_pbh = 4.3e-3
    f = 0.85*f_pbh 
    return (3.9e6 * f**2 *((t/t_0)**(-34/37))* (f**2 + σ_eq**2)**(-21/74) * min((P(m_1))/m_1, (P(m_2))/m_2) * ((P(m_1))/m_1 + (P(m_2))/m_2) * (m_1*m_2)**(3/37) * (m_1+m_2)**(36/37))
    
    
Merge = np.zeros((10, 10))


for i in range(10):
    for j in range(10):
        if i==j:
            Merge[i:(i+1), j:(j+1)] = 0.5*Merger_power(m_1_list[i], m_2_list[j],t)*Δ_1_list[i]*Δ_2_list[j]
            
        else:
            Merge[i:(i+1), j:(j+1)] = Merger_power(m_1_list[i], m_2_list[j],t)*Δ_1_list[i]*Δ_2_list[j]
            



mask = np.tril(Merge,k=-1)  #created an array of shape Merger_12 with all the elements in it being 0.
fig,ax= plt.subplots(figsize=(8, 6))  

ax = sns.heatmap(Merge, mask=mask,  cmap='viridis', cbar=True,
                 cbar_kws={"shrink": 0.85,'label': ' $Merger\hspace{0.5}Rate \hspace{0.5}$ [Gpc$^{-3}$ yr$^{-1}$]'},
                 vmin=0, xticklabels = m_1_list, yticklabels = m_2_list)


# make frame visible
for _, spine in ax.spines.items():
    spine.set_visible(True)


    
ax.invert_yaxis()
plt.xlabel('$M_{1}\hspace{0.5}[M_{☉}]$')
plt.ylabel('$M_{2}\hspace{0.5}[M_{☉}]$')
plt.title('Powerlaw PBHs mass distribution for same Δ')
plt.savefig('Powerlaw PBHs mass Distribution for same Δ.pdf')
plt.show()     
```

```python
# Calculation of the 2D merger rate for Log-Normal PBH mass Distribution. 
#The resolution of the PBHs masses m_1 is Δ_1 and of m_2 is Δ_2 with Δ_1=Δ_2=Δ.



def Q(m):               #Log-Normal PBH mass Distribution
    σ=0.6
    m_c=15               #in units of M☉
    return  math.exp(-((math.log(m/m_c))**2)/(2*(σ**2)))/(σ*m*math.sqrt(2*π))




def Merger_lognormal(m_1, m_2, t):
    f_pbh = 4.3e-3
    f = 0.85*f_pbh 
    return (3.9e6 * f**2 *((t/t_0)**(-34/37))* (f**2 + σ_eq**2)**(-21/74) * min((Q(m_1))/m_1, (Q(m_2))/m_2) * ((Q(m_1))/m_1 + (Q(m_2))/m_2) * (m_1*m_2)**(3/37) * (m_1+m_2)**(36/37))
    
    
Merger = np.zeros((10, 10))


for i in range(10):
    for j in range(10):
        if i==j:
            Merger[i:(i+1), j:(j+1)] = 0.5*Merger_lognormal(m_1_list[i], m_2_list[j],t)*Δ_1_list[i]*Δ_2_list[j]
            
        else:
            Merger[i:(i+1), j:(j+1)] = Merger_lognormal(m_1_list[i], m_2_list[j],t)*Δ_1_list[i]*Δ_2_list[j]
            



mask = np.tril(Merger,k=-1)  #created an array of shape Merger_12 with all the elements in it being 0.
fig,ax= plt.subplots(figsize=(8, 6))  

ax = sns.heatmap(Merger, mask=mask,  cmap='viridis', cbar=True,
                 cbar_kws={"shrink": 0.85,'label': ' $Merger\hspace{0.5}Rate \hspace{0.5}$ [Gpc$^{-3}$ yr$^{-1}$]'},
                 vmin=0, xticklabels = m_1_list, yticklabels = m_2_list)


# make frame visible
for _, spine in ax.spines.items():
    spine.set_visible(True)


    
ax.invert_yaxis()
plt.xlabel('$M_{1}\hspace{0.5}[M_{☉}]$')
plt.ylabel('$M_{2}\hspace{0.5}[M_{☉}]$')
plt.title('Log-Normal PBH mass Distribution for same Δ')
plt.savefig('Log-Normal PBH mass Distribution for same Δ.pdf')
plt.show()     
```

```python
#The resolution of the PBHs masses m_1 is Δ_1 and of m_2 is Δ_2.

M_1=6
M_2=8

Y_1=math.log(M_1)
Y_2=math.log(M_2)
dY=(Y_2-Y_1)


Y_list=np.zeros(8)

for i in range(len(Y_list)):
    Y_list[i]= (Y_1+(i*dY))
    

M_1_list=np.zeros(len(Y_list))
for i, Y in enumerate(Y_list):
    M_1_list[i]=(math.ceil(math.exp(Y)))
    
    

Δ_1_list=np.zeros(len(Y_list))

for i, Y in enumerate(Y_list):
    Δ_1_list[i] = math.exp(Y + 0.5*dY)-math.exp(Y - 0.5*dY)
     


        
Z_list=np.zeros(8)

for i in range(len(Z_list)):
    Z_list[i]= (Y_2+((i+1)*dY))
    

M_2_list=np.zeros(len(Z_list))
for i, Z in enumerate(Z_list):
    M_2_list[i]=(math.ceil(math.exp(Z)))
    

dZ=(Z_list[1] - Z_list[0])

Δ_2_list=np.zeros(len(Z_list))

for i, Z in enumerate(Z_list):
    Δ_2_list[i] = math.exp(Z + 0.5*dZ) - math.exp(Z - 0.5*dZ)
     
        
m_1_List=[int(M_1) for M_1 in M_1_list]
m_2_List=[int(M_2) for M_2 in M_2_list]


print("m_1=",m_1_List,"M☉")
print("m_2=",m_2_List,"M☉")
print("Δ_1=",Δ_1_list,"M☉")
print("Δ_2=",Δ_2_list,"M☉")
```
