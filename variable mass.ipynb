```python
# Solution of eq.(12) mentioned in paper "Black Holes’ Dark Dress: On the merger rate of a subdominant population of primordial black holes".

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.integrate import odeint
def f(u,t):
    return [u[1], ((-G*M/(u[0]**2)) - (u[0]/(4*t**2)))]
G= 6.67e-11
M= (1.989e32)
r0= [1e12,2.12044e6]
ts= np.linspace(2.358e5,8.465e9,2000)
us= odeint(f,r0,ts)
rs= us[:,0]
r_max= (np.max(rs))
t_max = ts[np.argmax(rs)]
print('r_ta=',r_max,'m')
plt.plot(ts,rs,'m')
plt.xlabel('t in seconds')
plt.ylabel('r in meters')
plt.title(' r = 10e12m & M = 100 solar mass')
plt.show()
```
