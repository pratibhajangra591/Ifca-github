```python
# I tried to plot eq.(9) given in the paper "The merger rate of primordial-black-hole binaries" for  λ<<1.

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.integrate import odeint



def f(u,s):
    return [u[1], ((s*u[1]-u[0])/(s**2))-((u[0]*(s**2))/(λ*((u[0]**2+b**2)**(3/2))))]
separation0= [1e-12,1]
λ=0.009
b=1e-4
ss= np.linspace(1e-12,1e-2,6000) # I tried all combinations with the initial and final value of the scale factor with the variation of the impact parameter
                                 # but got the correct behaviour in this range.
us= odeint(f,separation0,ss)
separations= us[:,0]
As=ss/λ
Bs=separations/λ



def j(n,p):
    return [n[1], ((p*n[1]-n[0])/(p**2))-((n[0]*(p**2))/(λ*((n[0]**2+c**2)**(3/2))))]
z0= [1e-12,1]
λ=0.009
c=1e-5
ps= np.linspace(1e-12,1e-2,6000) 
ns= odeint(j,z0,ps)
zs= ns[:,0]
Ps=ps/λ
Zs=zs/λ

def g(v,t):
    return [v[1], ((t*v[1]-v[0])/(t**2))-((v[0]*(t**2))/(λ*((v[0]**2+d**2)**(3/2))))]
λ=0.009
d= 1e-6
x0= [1e-12,1]
ts= np.linspace(1e-12,1e-2,6000) 
vs= odeint(g,x0,ts)
xs= vs[:,0]
Ts=ts/λ
Xs=xs/λ


def h(m,k):
    return [m[1],((k*m[1]-m[0])/(k**2))-((m[0]*(k**2))/(λ*((m[0]**2+e**2)**(3/2)))) ]
λ=0.009
e= 1e-7                                       # I tried for impact parameter = 1e-8 also, but in that case as the value is really small so we require around 
                                              # 600000000 no. of points to see the expected bahaviour, as there should be a limit on how small the impact 
                                                #  parameter can be I tried to avoid that behaviour of the binary for for such values. 
    
y0= [1e-12,1]
ks= np.linspace(1e-12,1e-2,6000) 
ms= odeint(h,y0,ks)
ys= ms[:,0]
Ks=ks/λ
Ys=ys/λ



plt.figure(figsize=(18, 12))

plt.plot(As,Bs,'b--')
plt.plot(Ps,Zs,'m')
plt.plot(Ts,Xs,'r')
plt.plot(Ks,Ys,'g')

plt.xlabel('s/λ')
plt.ylabel('X/λ')  
plt.legend(["b=1e-4","b=1e-5","b=1e-6","b=1e-7"])     
plt.show()

# Please check and let me know if this is correct or not.
```
