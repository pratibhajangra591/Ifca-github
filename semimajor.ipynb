```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.integrate import odeint
def f(u,b):
    return [u[1], -((u[0]*(b**2))/(λ*((u[0]**2+a**2)**(3/2))))] # with 2nd part of the eq.(9)
X0= [0,1]
λ=1 # Here if I am using λ much less than 1, some numerical error is coming. 

a=1e-2 # Also, If I am using some positive value of a then straight line behaviour is coming.

bs= np.linspace(0.001,2.0,40)# this range I fixed after hit and trial and taking hint from the paper.

us= odeint(f,X0,bs)
Xs= us[:,0]
As=bs/λ
Bs=Xs/λ
plt.plot(As,Bs,'r')
plt.xlabel('b/λ')
plt.ylabel('X/λ')
plt.title(' ')
plt.show()
```
