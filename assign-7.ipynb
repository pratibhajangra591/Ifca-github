```python
#1a.) solution of the full equation B2 in the tau interval, tau_initial=2.39974e8 to tau_final=2.27714e11 on linear scale

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.integrate import odeint
def f(u,x):
    return [u[1], ((-1/(2*u[0]**2)) - (u[0]/(4*x**2)))]
y0= [3.39197e6,7.06737e-3]
xs= np.linspace(2.39974e8,2.27714e11,20)
us= odeint(f,y0,xs)
ys= us[:,0]
plt.plot(xs,ys,'r')
plt.xlabel('t/2GM')
plt.ylabel('r/2GM')
plt.title('using r = 10e12m & M = 100 solar mass')
plt.show()
```
