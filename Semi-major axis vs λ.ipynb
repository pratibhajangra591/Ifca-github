```python
# Fitting of the the curve for the value of semi-major axis, a/ vs  for the PBH binary without 
# dark matter halos.


x = np.array(_list)

y = np.array(a_list)  


z = np.polyfit(x, y, 1)
p = np.poly1d(z)
print("p=",p)


plt.figure(figsize=(8,5))
plt.yscale('log')
plt.loglog(x,y,'go')
plt.plot(x,p(x),'r')
plt.xlabel('$$',fontsize=13)
plt.ylabel('$a/(m)$',fontsize=13)
plt.show()

#This gives semi-major axis, a=(0.09647+0.01652^2)m.




#Calculation of the Percentage error in the numerical and fitted value of semi-major axis, a/ 
#for PBH binary without dark matter halos.

y_list = y.tolist()
q= np.array(p(x))
q_list=q.tolist()

def percentage_error(numerical, fitted): 
    return (np.abs((numerical - fitted) / numerical)) * 100



error_list=np.zeros(len(_list))

for i, (y,q) in enumerate(zip(y_list,q_list)):
    error_list[i] = percentage_error(y, q)    #Percentage error 


error_max=np.max(error_list)
print("Maximum Percentage error= "'%0.4f'%error_max,"%")
```
